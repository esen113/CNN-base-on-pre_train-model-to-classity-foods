```python
import tensorflow as tf
#shape
img_height = 224
img_width = 224
batch_size = 16

#split the data
validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/drive/MyDrive/validation",
    image_size=(img_height, img_width),
    batch_size=batch_size,
    label_mode='int',
    shuffle=False
)

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/drive/MyDrive/labeled",
    image_size=(img_height, img_width),
    batch_size=batch_size,
    label_mode='int',
    shuffle=True
)
```

```python
from tensorflow.keras.layers import RandomRotation, RandomZoom, RandomFlip

# augmentation
data_augmentation = tf.keras.Sequential([
    RandomRotation(0.1),
    RandomZoom(0.1),
    RandomFlip("horizontal"),
])

# preprocess
def preprocess_image(image, label):
    image = tf.cast(image, tf.float32) / 255.0  # 归一化像素值
    return image, label

#
train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x, training=True), y))
train_dataset = train_dataset.map(preprocess_image)
train_dataset = train_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

#
validation_dataset = validation_dataset.map(preprocess_image)
validation_dataset = validation_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
import numpy as np
from tensorflow.keras.applications import EfficientNetB7
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import BatchNormalization
# base model
base_model = EfficientNetB7(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True
#train_layer = 3
for layer in base_model.layers[:-30]:
    layer.trainable = False
from tensorflow.keras.callbacks import ReduceLROnPlateau

# reduce learning rate
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',  # monitor
    factor=0.5,          # lower 0.5 every time
    patience=3,          # waite time
    min_lr=1e-6          # set boundry
)

from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
# trainable model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu',kernel_regularizer=l2(0.01)),
    BatchNormalization(),
    Dropout(0.6),
    Dense(11, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_dataset,
                    epochs=30,
                    validation_data=validation_dataset,
                    callbacks=[reduce_lr,early_stopping])
```
